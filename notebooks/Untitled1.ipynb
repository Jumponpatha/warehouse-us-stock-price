üè∑ Field-Level Metadata
1. Date

Type: timestamp with timezone

Description: Trading date and time for each price record.

Example: 2025-10-27 04:00:00+00:00

2. Open

Type: float

Description: Opening stock price at the beginning of the trading day.

Example: 189.99

3. High

Type: float

Description: Highest traded price during the day.

Example: 192.00

4. Low

Type: float

Description: Lowest traded price during the day.

Example: 188.42

5. Close

Type: float

Description: Closing stock price at the end of the trading session.

Example: 191.49

6. Volume

Type: bigint

Description: Total number of shares traded during the day.

Example: 153452700

7. Dividends

Type: float

Description: Dividend value paid on that date (per share).
Usually 0.0 on non-dividend days.

Example: 0.0

8. Stock Splits

Type: float

Description: Ratio representing split events.
0.0 means no split; e.g., 2.0 means a 2-for-1 split.

Example: 0.0

9. Symbol

Type: varchar

Description: The company ticker symbol representing the stock.

Example: NVDA

10. Ingested_Time

Type: timestamp with timezone

Description: Timestamp indicating when the data was ingested by the ETL pipeline.

Example: 2025-11-25 14:56:04.196316+00:00

```python
def get_sp500_list():
    # Get the list of S&P 500 companies from a reliable source
    url = "https://www.slickcharts.com/sp500"
    df = pd.read_html(url)[0]
    df['Symbol'] = df['Symbol'].str.replace('.', '-', regex=False)  # fix BRK.B ‚Üí BRK-B
    sp500_list = df['Symbol'].values.tolist()
    print("S&P 500 company list extracted successfully.")
    print(f"Total companies in S&P 500: {len(sp500_list)}")
    return sp500_list

# try:
#     # Get S&P 500 company list
#     sp500_list = get_sp500_list()

#     # Initialize an empty list to store data
#     data_list = []

#     for symbol in sp500_list:
#         ticker = yf.Ticker(symbol)
#         info = ticker.info  # Full metadata

#         data_list.append({
#                 "Exchange": info.get("exchange"),
#                 "Symbol": symbol,
#                 "Shortname": info.get("shortName"),
#                 "Longname": info.get("longName"),
#                 "Sector": info.get("sector"),
#                 "Industry": info.get("industry"),
#                 "Current_Price": info.get("currentPrice"),
#                 "Market_Cap": info.get("marketCap"),
#                 "Net_Income_To_Common": info.get("netIncomeToCommon"),
#                 "Ebitda": info.get("ebitda"),
#                 "Revenue_Growth": info.get("revenueGrowth"),
#             })

#         sp_profile_df = pd.DataFrame(data_list)
# except Exception as e:
#     print(f"The ERROR is: {e}")


# S&P 500 index ticker on Yahoo Finance
try:
    # Get S&P 500 company list
    sp500_list = get_sp500_list()

    # Initialize an empty list to store data
    data_list = []

    for symbol in sp500_list:
        ticker = yf.Ticker(symbol)
        hist = ticker.history(period='1mo', interval='1d', start=None, end=None)  # Historical price data

        hist.reset_index(inplace=True)
        hist['Symbol'] = symbol  # Add symbol column

        data_list.append(hist)

    sp_price_history_df = pd.concat(data_list, ignore_index=True)
except Exception as e:
    print(f"The ERROR is: {e}")
```
